#Tokenization

```python
import re
def tokenize(txt):
  tokens = re.split('\W+', txt)
  return tokens

df['ingredients_tokenized']= df['ingredients_clean'].apply(lambda x: tokenize(x.lower()))
df.head()
```

#Remove stop words

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
new_words=['text','cup','teaspoon','cups','page','ounce','ounces','see','note','about','c','tbsp','cut','sliced','green','red','minced','american','grated','elbow','cubed','mashed','homestyle','kosher','quartered','boiling','thawed','package']
for i in new_words:
    stopwords.append(i)
stopwords[30:]

```

```python
def remove_stopwords(txt_tokenized):
  txt_clean = [word for word in txt_tokenized if word not in stopwords ]
  
  digit=[word for word in txt_clean if word.isdigit()==False  ]
  return digit
df['ingredients_no_sw']= df['ingredients_tokenized'].apply(lambda x: remove_stopwords(x))
df.head()
```

#Get recommendatin by title using ingredients similarity

```python
def give_rec(title, dj=dj):
  idx = indices[title]
  dj_scores=list(enumerate(dj[idx]))
  dj_scores=sorted(dj_scores,key=lambda x: x[1], reverse=True)
  dj_scores=dj_scores[1:11]
  recipe_indices = [i[0] for i in dj_scores]
  return df['title'].iloc[recipe_indices]
  
```
